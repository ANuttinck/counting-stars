```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
# Read image
filename = '..\star_wars.png'
im = cv2.imread(filename)

# Show keypoints
plt.figure(figsize=(20, 16))
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
```

```python
# Setup SimpleBlobDetector parameters
params = cv2.SimpleBlobDetector_Params()

# Threshold parameters
params.minThreshold = 10  # 22
params.maxThreshold = 24
params.thresholdStep = 1
params.minDistBetweenBlobs = 1 # 4

# Color filter parameters
params.filterByColor = 1
params.blobColor = 255

# Size filter parameters
params.filterByArea = 1
params.minArea = 0.1 # 0.5
params.maxArea = 20  # 14

# Shape filter parameters
params.filterByConvexity = 0
# params.minConvexity = 0.8
params.filterByCircularity = 0
params.filterByInertia = 0
```

```python
# Set up the detector with the parameters above
detector_th10_24 = cv2.SimpleBlobDetector_create(params)

# Detect blobs
keypoints = detector_th10_24.detect(im)
```

```python
# printing parameter of the detector
for param in params.__dir__():
    if param[:2] != "__":
        print(param, str(getattr(params, param)))
print("Number of keypoints: ", len(keypoints))

# Draw detected blobs as red circles
# cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS ensures
# the size of the circle corresponds to the size of blob
im_with_keypoints = cv2.drawKeypoints(im, keypoints, np.array([]), (0, 0, 255),
                                      cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
# Show keypoints on the image
plt.figure(figsize=(20, 16))
plt.imshow(cv2.cvtColor(im_with_keypoints, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
ret, thresh1 = cv2.threshold(im, 23, 255, cv2.THRESH_BINARY)
plt.figure(figsize=(20, 16))
plt.imshow(thresh1,'gray')
```

```python
params_th24 = params

# Threshold parameters
params_th24.minThreshold = 23
params_th24.maxThreshold = 25
params_th24.thresholdStep = 1

detector_th24 = cv2.SimpleBlobDetector_create(params_th24)

# Detect blobs
keypoints_th24 = detector_th24.detect(im)
print("Number of keypoints with threshold: " + str(len(keypoints_th24)))

im_with_keypoints_th24 = cv2.drawKeypoints(im, keypoints_th24, np.array([]), (0, 0, 255),
                                              cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=(20, 16))
plt.imshow(cv2.cvtColor(im_with_keypoints_th24[300:500, 300:], cv2.COLOR_BGR2RGB))
```
